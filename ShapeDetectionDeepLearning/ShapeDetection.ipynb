```python
import tensorflow as tf
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from random import shuffle
from tensorflow.keras.layers import Conv2D, Flatten, MaxPooling2D, Dense, Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import to_categorical


import cv2

from sklearn.model_selection import train_test_split
# offline.init_notebook_mode(connected = True)
```

```python
folder = './shapes'
shapes = os.listdir(folder)


data = []
target = []
for i, shape in enumerate(shapes):
    path = os.path.join(folder, shape)
    for filename in os.listdir(path):
        image = cv2.imread(os.path.join(path , filename),0)
        image = image.reshape(200,200,1)
        data.append(image)
        target.append(i)
```

```python
len(target)
```

```python
imgplot = plt.imshow(data[0])
```

```python
data = np.array(data)
target = np.array(target)
target = to_categorical(target)
```

```python
imgplot = plt.imshow(data[4000])
```

```python
print(data.shape)
print(target.shape)
target[4000]
```

```python
X_full, X_test, y_full, y_test = train_test_split(data, target, stratify=target, random_state=123, test_size=0.25)
X_train, X_val, y_train, y_val = train_test_split(X_full, y_full, stratify=y_full, random_state=123, test_size=0.15)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_train = X_train / 255.0
X_val = X_val / 255.0
X_test = X_test / 255.0
```

```python
model = Sequential([
    Conv2D(32, (3,3),strides = 2 ,input_shape = (200,200,1), activation = 'relu'),
    MaxPooling2D(),
    Conv2D(32, (3,3), activation = 'relu'),
    MaxPooling2D(),
    Conv2D(32, (3,3), activation = 'relu'),
    MaxPooling2D(),
    Conv2D(32, (3,3), activation = 'relu'),
    MaxPooling2D(),
    Flatten(),
    Dropout(0.5),
    Dense(128, activation = 'relu'),
    Dropout(0.5),
    Dense(64, activation = 'relu'),
    Dropout(0.5),
    Dense(4, activation = 'softmax')
])
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

history = model.fit(X_train, y_train, epochs = 5, validation_data = (X_val, y_val))
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print('Accuracy is {}%'.format(accuracy*100))
```

```python
model.save('shape_detection_model')
```
